Veri Görselleştirme - Matplotlib - Seaborn

```python
#kategorik değişken : sütun grafik - countplot bar
#sayısal değişken : histogram, boxplot
```

```python
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns',None)
pd.set_option('display.width',500)
df = sns.load_dataset("titanic")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df["sex"].value_counts().plot()
```

```python
df["sex"].value_counts().plot(kind="bar")
```

Sayısal Değişken Görselleştirme

```python
plt.hist(df["age"])
plt.show()
```

```python
plt.boxplot(df["fare"])
```

Matplotlib Özellikleri

```python
import pandas as pd
import seaborn as sns
import numpy as np
pd.set_option('display.max_columns',None)
pd.set_option('display.width',500)
```

```python
#plot

x = np.array([1,8])
```

```python
y = np.array([0,150])
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
plt.plot(x,y,'o')
plt.show()
```

```python
x = np.array([2,4,6,8,10])
y = np.array([1,3,5,7,9])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'o')
plt.show()
```

```python
#marker

y = np.array([13,28,11,100])
```

```python
plt.plot(y,marker='o')
```

```python
plt.plot(y,marker='*')
```

```python
plt.plot(y,marker='d')
```

```python
plt.plot(y,marker='deniz')
```

```python
#line
y = np.array([13,28,11,100])
plt.plot(y,linestyle="dotted")
```

```python
#line
y = np.array([13,28,11,100])
plt.plot(y,linestyle="dashdot")
```

```python
#line
y = np.array([13,28,11,100])
plt.plot(y,linestyle="dotted",color="r")
```

```python
#multiple lines
x = np.array([23,18,31,10])
y = np.array([13,28,11,100])
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
#labels
x = np.array([80,85,90,95,100])
y = np.array([240,250,260,270,280])

plt.title("Bu ana başlık")
plt.xlabel("X elemanları")
plt.ylabel("Y bileşenleri")
plt.grid()

plt.plot(x,y)
plt.show()
```

```python
#subplots
x = np.array([80,85,90,95,100])
y = np.array([240,250,260,270,280])
plt.subplot(1,2,1)
plt.title("2")
plt.plot(x,y)
plt.show()
```

```python
#subplots
x = np.array([80,85,90,95,100])
y = np.array([240,250,260,270,280])
plt.subplot(1,3,1)
plt.title("2")
plt.plot(x,y)
plt.show()
```

```python
#subplots
x = np.array([80,85,90,95,100])
y = np.array([240,250,260,270,280])
plt.subplot(1,3,2)
plt.title("2")
plt.plot(x,y)
plt.show()
```

Seaborn

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df["sex"].value_counts()
```

```python
sns.countplot(x=df["sex"],data=df)
```

```python
df["sex"].value_counts().plot(kind="bar")
```

```python
#sayısal değişkenleri görselleştirme
```

```python
sns.boxplot(x=df["total_bill"])
plt.show()
```

```python
df["total_bill"].hist()
plt.show()
```
