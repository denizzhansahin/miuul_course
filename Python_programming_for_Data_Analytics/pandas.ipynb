Topluluaştırma ve Gruplaştırma - Aggregation / Grouping

```python
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns',None)
df = sns.load_dataset("titanic")
df.head()
```

```python
df["age"].mean()
```

```python
df.groupby("sex")["age"].mean()
```

```python
df.groupby("sex").agg({"age":"mean"})
```

```python
df.groupby("sex").agg({"age":["mean","sum"]})
```

```python
df.groupby("sex").agg({"age":["mean","sum"],
                       "embark_town":"count"})
```

```python
df.groupby("sex").agg({"age":["mean","sum"],
                       "survived":"mean"})
```

```python
df.groupby(["sex","embark_town"]).agg({"age":["mean"],
                       "survived":"mean"})
```

```python
df.head()
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age":["mean","sum"],
                       "survived":"mean"})
```

```python
df.groupby(["sex","embark_town","class"]).agg({
    "age":["mean","sum"],
                       
    "survived":"mean",
    "sex":"count"})
```

Apply / Lambda

```python
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns',None)
df = sns.load_dataset("titanic")
df.head()
```

```python
df["age2"] = df["age"]*2
df["age3"] = df["age"]*5
```

```python
df.head()
```

```python
df["age"]/10
```

```python
(df["age"]/10).head()
```

```python
for col in df.columns:
    if "age" in col:
        print(col)
```

```python
for col in df.columns:
    if "age" in col:
        print((df[col]/10).head())
```

```python
for col in df.columns:
    if "age" in col:
        df[col] = df[col]/10
```

```python
df.head()
```

```python
df[["age","age2","age3"]].apply(lambda x:x/10).head()
```

```python
df.loc[:,df.columns.str.contains("age")].apply(lambda x:x/10).head()
```

```python
df.loc[:,df.columns.str.contains("age")].apply(lambda x:(x-x.mean()/x.std())).head()
```

```python
def standart_scaler(col_name):
    return (col_name-col_name.mean()/col_name.std())
```

```python
df.loc[:,df.columns.str.contains("age")].apply(standart_scaler).head()
```

```python
df.loc[:,["age","age2","age3"]]=df.loc[:,df.columns.str.contains("age")].apply(standart_scaler).head()
```

```python
df.loc[:,df.columns.str.contains("age")]=df.loc[:,df.columns.str.contains("age")].apply(standart_scaler).head()
```

```python
df.head()
```

Birleştirme Join İşlemleri

```python
import numpy as np
import pandas as pd
m = np.random.randint(1,30,size=(5,3))
```

```python
df1 = pd.DataFrame(m,columns=["var1","var2","var3"])
```

```python
df2 = df1 + 99
```

```python
df1
```

```python
df2
```

```python
#alt alta birleştirme
pd.concat([df1,df2])
```

```python
#alt alta birleştirme
pd.concat([df1,df2],ignore_index=True)
```

```python
#yan yana birleştirme
pd.concat([df1,df2],ignore_index=True, axis=1)
```

```python
#yan yana birleştirme
pd.concat([df1,df2],ignore_index=False, axis=1)
```

```python
#merge ile birleştirme
df1 = pd.DataFrame({'employees': ['john', 'dennis', 'mark', 'maria'],
                    'group': ['accounting', 'engineering', 'engineering', 'hr']})

df2 = pd.DataFrame({'employees': ['mark', 'john', 'dennis', 'maria'],
                    'start_date': [2010, 2009, 2014, 2019]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on="employees")
```

```python
df3 = pd.merge(df1,df2,on="employees")
```

```python
df4 = pd.DataFrame({'group': ['accounting', 'engineering', 'hr'],
                    'manager': ['Caner', 'Mustafa', 'Berkcan']})
```

```python
#amacımız her çalışanın müdür bilgisine eriş
pd.merge(df3,df4)
```
